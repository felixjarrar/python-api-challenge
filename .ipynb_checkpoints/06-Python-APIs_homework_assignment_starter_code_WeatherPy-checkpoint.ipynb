```python
# Import
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# API 
from api_keys import weather_api_key

# import citipy
from citipy import citipy

# Output File
output_data = "cities.csv"

# latitudes and longitudes ranges
latitude_ranges = (-90, 90)
longtiude_ranges = (-180, 180)
```

## Generate Cities List

```python
# List 
latitudes_longitudes = []
cities = []

# random lat and lng combinations
latitudes = np.random.uniform(low=-90.000, high=90.000, size=1500)
longitutdes = np.random.uniform(low=-180.000, high=180.000, size=1500)
latitudes_longitudes = zip(latitudes, longitutdes)

# Find the nearest city for each latitude, longitude combination
for latitudes_longitudes in latitudes_longitudes:
    city = citipy.nearest_city(latitudes_longitudes[0], latitudes_longitudes[1]).city_name
    if city not in cities:
        cities.append(city)

# Print 
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Start API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 

# List 
city_data = []

# Print 
print("Beginning Data Retrieval NOW   ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop 
for i, city in enumerate(cities):
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0
    city_url = url + "&q=" + city
    
    print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))

    record_count += 1

    # Run API request 
    try:
        
        city_weather = requests.get(city_url).json()
        city_lats = city_weather["coord"]["lat"]
        city_lngs = city_weather["coord"]["lon"]
        city_max_temperatures = city_weather["main"]["temp_max"]
        city_all_humidity = city_weather["main"]["humidity"]
        city_all_clouds = city_weather["clouds"]["all"]
        city_wind_speeds = city_weather["wind"]["speed"]
        city_country_name = city_weather["sys"]["country"]
        city_dates = city_weather["dt"]

        # Append 
        city_data.append({"City": city, 
                          "Lat": city_lats, 
                          "Lng": city_lngs, 
                          "Max Temp": city_max_temperatures,
                          "Humidity":  city_all_humidity,
                          "Cloudiness":  city_all_clouds,
                          "Wind Speed": city_wind_speeds,
                          "Country": city_country_name,
                          "Date": city_dates})

    # In case of an error
    except:
        print("Sorry! The city was not found. We are skipping it...")
        pass
              
# Indicate loading is complete 
print("-----------------------------")
print("The Data Retrieval is Complete      ")
print("-----------------------------")
        
```

```python
# Convert JSONs into Pandas
city_data_pd = pd.DataFrame(city_data)

# Extract relevant info
latitudes = city_data_pd["Lat"]
max_temperatures = city_data_pd["Max Temp"]
the_humidity = city_data_pd["Humidity"]
the_cloudiness = city_data_pd["Cloudiness"]
wind_speeds = city_data_pd["Wind Speed"]

# Export into a csv
city_data_pd.to_csv(output_data, index_label="City_ID")

# Count
city_data_pd.count()
```

## Linear Regression

```python
# create function for Linear Regression 
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    
    # Run regresson on southern hemisphere
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    # Plot
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,text_coordinates,fontsize=15,color="red")
    plt.xlabel('Latitudes')
    plt.ylabel(title)
    print(f"The r-squared for the calculation is: {rvalue}")
    plt.show()
```

```python
# Northern Hemisphere and Southern Hemisphere DataFrames
northern_hemisphere_df = city_data_pd.loc[(city_data_pd["Lat"] >= 0)]
southern_hemisphere_df = city_data_pd.loc[(city_data_pd["Lat"] < 0)]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Northern Hemisphere linear regression
x_values = northern_hemisphere_df["Lat"]
y_values = northern_hemisphere_df["Max Temp"]
plot_linear_regression(x_values, y_values, 'Max Temperature',(6,30))
```
